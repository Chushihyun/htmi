```python
import numpy as np
import copy
```

```python
def train_data_process():
    txt = np.fromfile('hw6_train.dat',sep='\t', dtype=float)
    num=int(len(txt)/11)
    data=np.zeros((num,10))
    label=np.zeros((num))
    for i in range(num):
        data[i][:]=txt[i*11:i*11+10]
        label[i]=txt[i*11+10]
    return data,label

def test_data_process():
    txt = np.fromfile('hw6_test.dat',sep='\t', dtype=float)
    num=int(len(txt)/11)
    data=np.zeros((num,10))
    label=np.zeros((num))
    for i in range(num):
        data[i][:]=txt[i*11:i*11+10]
        label[i]=txt[i*11+10]
    return data,label

def data_combine(data,label):
    new_label=label.reshape(len(data),1)
    new_data=np.append(data,new_label,axis=1)
    return new_data
```

```python
def gini(data):
    total=1
    N=len(data)
    if N==0:
        return 0
    count1=0
    count2=0
    for i in range(N):
        if data[i][1]==1:
            count1+=1
        elif data[i][1]==-1:
            count2+=1
    total-=(count1/N)**2
    total-=(count2/N)**2
    
    return total
```

```python
def calculate_impurity(data1,data2):
    total=0
    total+=len(data1)*gini(data1)
    total+=len(data2)*gini(data2)
    return total
```

```python
def decision_stump(datas):
    index=np.argsort(datas[:,0])
    sorted_datas=datas[index]
    min_impurity=99999
    min_thres=-9999
    for i in range(len(datas)-1):
        impurity=calculate_impurity(sorted_datas[:i+1],sorted_datas[i+1:])
        if impurity<min_impurity:
            min_impurity=impurity
            min_thres=(sorted_datas[i,0]+sorted_datas[i+1,0])/2

    return min_impurity,min_thres
    
```

```python
def find_decision_stump(datas):
    min_impurity=99999
    min_dimension=-1
    min_thres=-100
    #print(datas.shape)
    for i in range(10):
        impurity,thres=decision_stump(datas[:,[i,10]])
        #print(impurity)
        #print(thres)
        if impurity < min_impurity:
            min_impurity = impurity
            min_dimension = i
            min_thres = thres
    return min_dimension,min_thres, min_impurity
```

```python
def data_split(threshold,datas,dimension):
    data1=[]
    data2=[]
    for i in range(len(datas)):
        if datas[i][dimension]<threshold:
            data1.append(datas[i])
        else:
            data2.append(datas[i])
    return np.array(data1),np.array(data2)
```

```python
def same_label(data):
    l=data[0][10]
    for i in range(len(data)-1):
        if data[i+1][10]!= l:
            return False
    return True
        
```

```python
class tree():
    def __init__(self, data):
        self.leaf=False
        if len(data)<0:
            self.leaf=True
            return
        
        elif same_label(data)==True:
            self.leaf=True
            self.label=data[0][10]
            
        else:
            self.dim, self.threshold, impurity= find_decision_stump(data)
            data1,data2=data_split(self.threshold,data,self.dim)
            self.left=tree(data1)
            self.right=tree(data2)
            
    def find(self, data):
        if self.leaf==True:
            return self.label
        else:
            if data[self.dim]<self.threshold:
                return self.left.find(data)
            else:
                return self.right.find(data)
```

```python
train_data, train_label = train_data_process()
test_data, test_label = test_data_process()
all_train_data=data_combine(train_data,train_label)
print(all_train_data.shape)
```

```python
def evaluate(tree,data,label):
    err=0
    for i in range(len(data)):
        ans=tree.find(data[i])
        if ans!=label[i]:
            err+=1
    output=err/len(data)
    return output
```

```python
def predict(tree,data):
    output=[]
    for i in range(len(data)):
        ans=tree.find(data[i])
        output.append(ans)
        
    new_output=np.array(output).reshape(len(data),1)
    return new_output
```

```python
def positive_negative(label):
    new_label=copy.deepcopy(label)
    for i in range(len(label)):
        if label[i][0]>=0:
            new_label[i]=1
        else:
            new_label[i]=-1
    return new_label
```
